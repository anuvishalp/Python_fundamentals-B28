```python
🧮 1. Calculator

✅ Python Version

def calculator(a, b, op):
    if op == '+': return a + b
    if op == '-': return a - b
    if op == '*': return a * b
    if op == '/': return a / b
    return "Invalid operator"

print(calculator(10, 5, '+'))

✅ PL/SQL Version

DECLARE
  a NUMBER := 10;
  b NUMBER := 5;
  op VARCHAR2(1) := '+';
  result NUMBER;
BEGIN
  CASE op
    WHEN '+' THEN result := a + b;
    WHEN '-' THEN result := a - b;
    WHEN '*' THEN result := a * b;
    WHEN '/' THEN result := a / b;
    ELSE DBMS_OUTPUT.PUT_LINE('Invalid operator');
  END CASE;

  DBMS_OUTPUT.PUT_LINE('Result: ' || result);
END;
/

✅ SQL Version (Using CASE)

SELECT
  CASE '+'
    WHEN '+' THEN 10 + 5
    WHEN '-' THEN 10 - 5
    WHEN '*' THEN 10 * 5
    WHEN '/' THEN 10 / 5
  END AS result
FROM dual;
======================================================================
🔐 2. Password Generator

✅ Python Version

import random
import string

def generate_password(length=12):
    chars = string.ascii_letters + string.digits + string.punctuation
    return ''.join(random.choice(chars) for _ in range(length))

print(generate_password())
-----------------------
✅ PL/SQL Version

DECLARE
  chars VARCHAR2(200) := 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789!@#$%^&*';
  pwd VARCHAR2(50);
  len NUMBER := 12;
BEGIN
  FOR i IN 1..len LOOP
    pwd := pwd || SUBSTR(chars, TRUNC(DBMS_RANDOM.VALUE(1, LENGTH(chars))), 1);
  END LOOP;

  DBMS_OUTPUT.PUT_LINE('Password: ' || pwd);
END;
/

❌ SQL Version
SQL is not suitable for password generation logic — PL/SQL handles it.
===============================================================
📝 3. To‑Do App

✅ Python Version

tasks = []

def add_task(t):
    tasks.append(t)

def list_tasks():
    for i, t in enumerate(tasks, 1):
        print(i, t)

add_task("Learn Python")
add_task("Practice SQL")
list_tasks()
---------------------------
✅ SQL Version (Table + Insert + Select)

CREATE TABLE todo (
  id NUMBER GENERATED ALWAYS AS IDENTITY,
  task VARCHAR2(200),
  status VARCHAR2(20) DEFAULT 'pending'
);

INSERT INTO todo (task) VALUES ('Learn Python');
SELECT * FROM todo;
-----------------------------
✅ PL/SQL Version

DECLARE
  v_task VARCHAR2(200) := 'Learn PL/SQL';
BEGIN
  INSERT INTO todo(task) VALUES (v_task);
  COMMIT;
END;
/
===============================================
🕸️ 4. Web Scraper

✅ Python Version (requests + BeautifulSoup)

import requests
from bs4 import BeautifulSoup

url = "https://example.com"
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

print(soup.title.text)

❌ SQL / PL/SQL-> Not suitable for web scraping.
================================================
🌐 5. API‑Based App

✅ Python Version

import requests

response = requests.get("https://api.github.com/users/octocat")
data = response.json()
print(data["login"], data["public_repos"])

❌ SQL / PL/SQL -> APIs are not supported directly (unless using UTL_HTTP in PL/SQL).
-----------------------------
PL/SQL Version (UTL_HTTP)

DECLARE
  req  UTL_HTTP.req;
  resp UTL_HTTP.resp;
  buffer VARCHAR2(4000);
BEGIN
  req := UTL_HTTP.begin_request('https://api.github.com/users/octocat');
  resp := UTL_HTTP.get_response(req);

  LOOP
    UTL_HTTP.read_line(resp, buffer, TRUE);
    DBMS_OUTPUT.PUT_LINE(buffer);
  END LOOP;

  UTL_HTTP.end_response(resp);
END;
/
==============================================
🎮 6. Simple Game (Snake / Pong)

✅ Python Version (Mini Snake Skeleton)

import turtle

win = turtle.Screen()
win.title("Snake Game")
win.bgcolor("black")

head = turtle.Turtle()
head.shape("square")
head.color("white")
head.penup()

while True:
    head.forward(20)

❌ SQL / PL/SQL -> Games cannot be built in SQL/PLSQL.
===============================================
⚙️ Automation Projects

📁 7. Rename Files

✅ Python Version

import os

folder = "C:/temp"
for i, f in enumerate(os.listdir(folder)):
    os.rename(os.path.join(folder, f), os.path.join(folder, f"file_{i}.txt"))

======================================================
📧 8. Send Emails

✅ Python Version

import smtplib
from email.mime.text import MIMEText

msg = MIMEText("Hello from Python!")
msg["Subject"] = "Test Email"
msg["From"] = "you@example.com"
msg["To"] = "friend@example.com"

with smtplib.SMTP("smtp.gmail.com", 587) as server:
    server.starttls()
    server.login("you@example.com", "password")
    server.send_message(msg)
=======================================================
📊 9. Process CSV / Excel

✅ Python Version (pandas)

import pandas as pd

df = pd.read_csv("data.csv")
df["total"] = df["price"] * df["qty"]
df.to_excel("output.xlsx", index=False)
--------------------------------
✅ SQL Version (Load CSV into table)

LOAD DATA INFILE 'data.csv'
INTO TABLE sales
FIELDS TERMINATED BY ','
(price, qty);
------------------------------
✅ PL/SQL Version (Process rows)

DECLARE
  CURSOR c IS SELECT price, qty FROM sales;
BEGIN
  FOR r IN c LOOP
    DBMS_OUTPUT.PUT_LINE('Total: ' || (r.price * r.qty));
  END LOOP;
END;
/
======================================
🕷️ 10. Scrape Websites (Automation)

✅ Python Version

import requests
from bs4 import BeautifulSoup

html = requests.get("https://example.com").text
soup = BeautifulSoup(html, "html.parser")

for link in soup.find_all("a"):
    print(link.get("href"))
```
